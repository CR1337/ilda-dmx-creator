```python
def sort_and_filter(lst, k):
    # Count the occurrences of each number in the list
    counts = Counter(lst)
    
    # Filter out numbers that occur less than k times
    filtered = [x for x in lst if counts[x] >= k]
    
    # Sort the filtered list by the frequency (descending), and then by the number itself
    sorted_list = sorted(set(sorted(filtered, key=lambda x: (-counts[x], x))))
    
    return sorted_list
```
